```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
mean = 3        # Średnia
deviation = 1     # Odchylenie standardowe

# Wygenerowanie danych
data_100 = np.random.normal(mean, deviation, 100)     
data_10000 = np.random.normal(mean, deviation, 10000) 

data_100[:30], data_10000[:30]


# def gauss(x, sigma=1, mu=3):
#     return stats.norm.pdf(x, mu, sigma)
# x = np.linspace(0, 100)
# x[:30]

#nie wiem dlaczego ta opcja nie działała?
```

```python
median_100 = np.median(data_100)
print(f'Mediana zbioru 100 to: {median_100}')
```

```python
mode_100 = stats.mode(data_100, keepdims=False)
print(f'Najwięcej wartości zbioru 100 to: {mode_100}')
```

```python
q0 = np.quantile(data_100, 0.0)
q1 = np.round(np.quantile(data_100, 0.25))
q3 = np.round(np.quantile(data_100, 0.75))
q4 = np.round(np.quantile(data_100, 1.0))
print(f'Wartość min to: {q0}')
print(f'Q1: {q1}')
print(f'Q3: {q3}')
print(f'Wartość max Q4 to: {q4}')
```

```python
iqr = q3-q1
print(f'Rozstęp międzykwartylowy wynosi: {iqr}')
```

```python
variance_100 = np.var(data_100, ddof=1)
print(f"Wariancja wynosi: {variance_100}")
```

```python
deviation_100 = variance_100**(1/2)
print(f'Odchylenie standardowe wynosi: {deviation_100}')
```

```python
plt.hist(data_100, bins=20)
plt.title("Statystyki dla zbioru 100")
plt.axvline(x=mean, color='red', label='Średnia')
plt.axvline(x=median_100, color='yellow', label='Mediana')
plt.axvline(x=mode_100[1], color='black', label='Moda')
plt.axvline(x=q0, color='pink', label='Min')
plt.axvline(x=q1, color='pink', label='Q1')
plt.axvline(x=q3, color='pink', label='Q3')
plt.axvline(x=q4, color='pink', label='Max')
plt.axvline(x=iqr, color='purple', label='IQR')
plt.axvline(x=variance_100, color='green', label='Variance')
plt.axvline(x=deviation_100, color='orange', label='Deviation')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xlabel('Zakres')
plt.ylabel('Ilość')
plt.show()
```

```python
scaler_standardized = StandardScaler()
standardized_data_100 = scaler_standardized.fit_transform(data_100.reshape(-1, 1))
data_100 = data_100.reshape(-1, 1) #to zgodnie z Twoimi uwagami, ale nie mam pewności czy o to chodziło
plt.hist(standardized_data_100, bins=20)
plt.axvline(x = standardized_data_100.mean(), color='red', label = 'Średnia')
plt.title('StandardScaler dla 100')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xlabel('Zakres')
plt.ylabel('Ilość')
plt.show()
print(f'Średnia to: {standardized_data_100.mean()}')
print(f'Odchylenie standardowe to: {np.std(standardized_data_100, ddof=1)}')
```

```python
scaler_MinMax = MinMaxScaler()
normalized_data_100 = scaler_MinMax.fit_transform(data_100.reshape(-1, 1))
data_100 = data_100.reshape(-1, 1)
plt.hist(normalized_data_100, bins=20)
plt.axvline(x = normalized_data_100.mean(), color='red', label = 'Średnia')
plt.title(f'MinMax Scaler dla 100')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xlabel('Zakres')
plt.ylabel('Ilość')
plt.show()
print(f'Średnia: {normalized_data_100.mean()}')
print(f'Odchylenie standardowe: {np.std(normalized_data_100, ddof=1)}')
```

```python
def find_outliers(data):
    Q1 = np.percentile(data, 25)
    Q3 = np.percentile(data, 75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data < lower_bound) | (data > upper_bound)]
    return outliers
outliers_100 = find_outliers(data_100)
outliers_10000 = find_outliers(data_10000)
print("Outliers w zbiorze 100:", outliers_100)
print("Outliers w zbiorze 10000:", outliers_10000)
```
